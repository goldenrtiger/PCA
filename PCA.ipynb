```python
import numpy as np

fy = lambda x: x + 2.3

n = 50
#-- y = x + 2.3
x = np.linspace( 0, 50, num = n); y = fy( x )
x += np.random.randn( n ) * 5; y += np.random.randn( n ) * 5

ps = np.zeros( (3, n) )
ps[0] = x
ps[1] = y
```

```python
import matplotlib.pyplot as plt

f = plt.figure()
ax = f.add_subplot(111)
x0 = 0;                         y0 = fy( x0 )
x1 = n;                         y1 = fy( x1 )
x = np.array( [x0, x1] );       y = np.array( [y0, y1] )
ax.scatter(ps[0], ps[1])
plt.show()

def plot2D( x_s, y_s, x_l, y_l ):
    f = plt.figure()
    ax = f.add_subplot(111)
    ax.scatter(x_s, y_s)
    #-- through PCA
    ax.plot( x_l, y_l, c='b' )
    #-- original
    ax.plot( x, y, c='y' )
    plt.show()
```

```python
from PCA import PCA

data = np.transpose( ps )
pca = PCA()
e = pca.process( data )
e = e / np.linalg.norm( e )
e
x0 = 0;     y0 = x0 + e
x1 = n;     y1 = x1 + e
x = np.array( [x0, x1] ); y = np.array( [y0, y1] )
plot2D( ps[0], ps[1], x, y)
plt.show() 

```
